# MatplotLib

Matplotlib is a plotting library for the python programming language and its numerical mathematics.

Some majar pros of Matplotlib are.

* Generally easy to get started for simple plots
* Support for custom labels and texts
* High-quality output in many formats
* very customizable in genral.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

# Analysis

* Univariate Analysis--> Analyzing single variable at a time.
* Bivariate Analysis--->Analyzing considering 2 variable at time.
* Multivariate Analysis-->Analyzing more than 2 variable at time.

# Histogram

 * In histogram x axis contains variable and y axis contains frquency of that variable.

Important Parameter in Histogram

* data 
* Bins(default=10)

```python
plt.hist(df["Grade"])
plt.xlabel("Grade")  #set label for the x-axis
plt.ylabel("Frequency")  #Set the label for the y axis
plt.title("Histogram")  #set title of plots.
plt.show()
```

# Box plot

* Use: to check, the given data has outliers or not.

```python
plt.boxplot(df["Grade"])
plt.title("Boxplot of Grade")
plt.show()
```

# Scatter plot

Important Parameters in Scatterplot.

* X-axis values
* y-axis values
* color(default=Blue)

```python
plt.scatter(df["StudyHours"],df["Grade"],color="red",marker="*",s=30)
```

# Stacked And Unstacked bar plot.

* Stacked bar: Comparing two bar in single bar.

* Unstacked bar: Comparing two bar side by side.
    

```python
company=["Google","Amazon","MSFt","FB"]
revenue=[90,136,89,27]
profit=[40,20,30,12]
plt.bar(company,revenue,label="Revenue")
plt.bar(company,profit,label="profit")
plt.legend()
plt.show()
```

```python

df=pd.DataFrame({"company":["Google","Amazon","MSFt","FB"],"revenue":[90,136,89,27],"profit":[40,20,30,12]})
```

```python
company=["Google","Amazon","MSFt","FB"]
revenue=[90,136,89,27]
profit=[40,20,30,12]
xpos=np.arange(len(company))
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="profit")
plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title("Us Technology Stocks")
plt.legend()
plt.show()
```

* Revenu wise--->amazon is higher
* Profit wise --->Google is higher


# Subplot

* Use to represent multiple bar chart in one image.

* the subplot() function take three argument that describe the layout of figure.

* first two argument for row and column and thired for index of the current plot.

```python
#Multiple bar cahrt on single image.

ecommerce=["snapdeal",'alibaba','amazon','flipkart']
Q1_profit=[45,100,90,40]
Q2_profit=[44,39,59,79]
Q3_profit=[54,65,34,89]
Q4_profit=[34,44,112,144]

#Creating different bar charts on one image using subplot() function.

plt.figure(figsize=(10,10))

#Creating bar chart in first cell of figure having 3 rows ,2 columns.

plt.subplot(3,2,1)
plt.bar(ecommerce,Q1_profit)
plt.title("Quarter 1 profit")


#Creating bar chart in second cell

plt.subplot(3,2,2)
plt.bar(ecommerece,Q2_profit)
plt.title("Quarter 2 Profit")

#creating bar chart in third cell

plt.subplot(3,2,3)
plt.bar(ecommerece,Q3_profit)
plt.title("Quarter 3 profit")

#creating bar chart in 6 th cell

plt.subplot(3,2,6)
plt.bar(ecommerece,Q4_profit)
plt.title("Quarter profit")

#Display the charts 
plt.show()
```

```python
# Import libraries
import requests
from bs4 import BeautifulSoup

# Make a request to the website
response = requests.get("http://www.maharera.maharashtra.gov.in/agents-search-result")

# Parse the response as HTML using Beautiful Soup
soup = BeautifulSoup(response.text, "html.parser")

# Find all elements with class name "card-title"
titles = soup.find_all(class_="")  



# Loop through the titles and print them as text
for title in titles:
    print(title.get_text())
```

```python
import requests
from bs4 import BeautifulSoup

url = 'http://www.maharera.maharashtra.gov.in/agents-search-result?agent_name=&agent_project_name=&agent_location=&agent_state=27&agent_division=&agent_district=&page=2&op=Search#main-content'
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

# Extract data using BeautifulSoup methods
# For example, if agents are listed in a certain class:
agents = soup.find_all('div', class_=["data-current-data"
])
print(agents)

# Process the extracted data as needed
for agent in agents:
    # Process and print or store the agent details
    print(agent.text)
```

Combining DataFrame

```python
temperature_df=pd.DataFrame({"city":["pune","mumbai","banglore","delhi"],"temperature":[32,23,34,33]})

temperature_df
```

```python
humidity_df=pd.DataFrame({"city":["pune","mumbai","chennai","rajasthan"],"humidity":[80,70,60,50]})

humidity_df
```

Option 1-->append

```python
df=temperature_df.append(humidity_df)
```

```python
df
```

Option 2-->Concatenate(Recomm)

```python
df=pd.concat([temperature_df,humidity_df])
df
```

```python
#Ignore_index--->first add two dataframe and then give index to this.

pd.concat([temperature_df,humidity_df],ignore_index=True)
```

```python
pd.concat([temperature_df,humidity_df],axis=1)
```

Merging of DataFrame

* merging is similar to join in sql.
* It is use to extract similar data from two table and create single table for easy reportig.
* 4 Types of Merge.

  1.Inner merge: Common record from both Dataframes.
  
  2.Outer merge: All common and different from both dataframes.
  
  3.left merge : All record from left side and  common from right side                  dataframe.
  
  4.Right merge: All record from right side and common from left side
                 DataFrame.
 

```python
#inner merge or join

pd.merge(temperature_df,humidity_df,how='inner',on="city")
```

```python
#Outer merge 

pd.merge(temperature_df,humidity_df,how='outer',on="city")
```

```python
#left merge

pd.merge(temperature_df,humidity_df,how='left',on="city")
```

```python
#Right Merge

pd.merge(temperature_df,humidity_df,how='right',on="city")
```
